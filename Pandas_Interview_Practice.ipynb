## create a new category "Investment_category" in the dataframe df which will classify countries into "High, "Medium" and "low"based on digital infrastructure values

```python
# define thresholds for investment categories

high_threshold = df["Digital_Infrastructure_Investment"].quantile(0.75) # top 25% as high

low_threshold = df["Digital_Infrastructure_Investment"].quantile(0.25)  # bottom 25% as low


# function to categorize each row

def categorize_investments(investment):
    if investment >= high_threshold:
        return "High"
    elif investment <= low_threshold:
        return "Low"
    else:
        return "Medium"


# creating a new column based on the above defined logic

df["Investment_category"] = df["Digital_Infrastructure_Investment"].apply(categorize_investments)
```

# Merge & Concatenation of Data Frames

## Merging

```python
# How to merge data frames to get a holistic view

"""

the "ON" Parameter specifies the key, similar to primary key in sql.

Note: all missing values are replaced with NaN.

"""
```

```python
# lets try to understand merge usage by simple example

d1 = {'Name': ['John', 'Alice', 'Maria'], 'Country': ['India', 'India', 'USA'], 'Role': ['CEO', 'CTO', 'CTO']}
df1 = pd.DataFrame(d1)
print('DataFrame 1:\n', df1)



df2 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['John', 'Kat', 'Jane']})
print('DataFrame 2:\n', df2)

# when we try to merge two data frames - by default it makes an INNER JOIN

# JOINS HERE ARE SIMILAR TO THE ONES LEARNT IN SQL

df_merged = df1.merge(df2)

print("Result:\n",df_merged)
```

## Merging data frames with left, righ and outer joins

```python
df_left_merged = df1.merge(df2,how = 'left')

df_left_merged
```

```python
df_right_merged = df1.merge(df2, how='right')

df_right_merged
```

```python
df_outer = df1.merge(df2, how='outer')

df_outer
```

```python
# simply it could be understood by terms. split, apply and combine. (similar to analogy of window functions in sql)


"""

Split: breaking up and grouping a data frame depending on the value of specified key.

Apply: computing some function, usually an aggregate, transformation or filtering within the indiviual groups.

Combine: merge the results of these operations into an output array.

"""
```
