## Q2: CUSTOMER ANALYSIS

PROBLEM STATEMENT:

- you are given a dataframe containing booking data from a hotel with columns: guest_Id, gender & date of booking

- create a new dataset with summkary of information on their bookings including columns: guest_id, gender, most_recent_booking, booking_count.

- most recent booking should diplay the date of guests most recent booking and booking count should display the count of bookings guest has made.

```python
# lets try to build logical approach first.

"""

    1. load the data
    2. convert dates.
    3. group and aggregate.
    4. create the summary.
    5. return the result.

"""
```

```python
bookings = pd.read_csv("BookingsData.csv")
```

```python
bookings
```

```python
bookings.info()
```

```python
# converting date of booking column to date time format

bookings["date_of_booking"] = pd.to_datetime(bookings["date_of_booking"],dayfirst=True)
```

```python
# group by guest id and gender to fetch most recent booking and count

summary_df = bookings.groupby(["guest_id","gender"]).agg(
    most_recent_booking = ("date_of_booking", "max"),
    booking_count = ("date_of_booking","count")

).reset_index()
```

Q3: ENROLLMENTS IN UNIVERSITY

```python
enrollments = pd.read_csv("enrollments.csv")
```

```python
enrollments
```

```python
enrollments["enrollment_date"] = pd.to_datetime(enrollments["enrollment_date"])
enrollments = enrollments.sort_values(by="enrollment_date")
```

```python
previous_enrollments = {}

def check_previous_enrollments(row):
    student_id = row["student_id"]
    department = row["department"]

    if student_id not in previous_enrollments:
        previous_enrollments[student_id] = set()
    if department in previous_enrollments[student_id]:
        return 1

    else:
        previous_enrollments[student_id].add(department)
        return 0


enrollments["department_previously_enrolled"] = enrollments.apply(check_previous_enrollments, axis=1)
```

```python
enrollments_final = enrollments[["course_name", "department_previously_enrolled"]]
```

```python
enrollments_final
```

Q4: USERS BY AVERAGE TIME

```python
user_activities = pd.read_csv("user_activities.csv")
```

```python
user_activities
```

```python
user_activities["timestamp"] = pd.to_datetime(user_activities["timestamp"], dayfirst=True)

login_df = user_activities[user_activities["action"] == "login"][["user_id","timestamp"]]

logout_df = user_activities[user_activities["action"] == "logout"][["user_id","timestamp"]]
```

```python
df = pd.merge(login_df, logout_df, on='user_id', suffixes=['_login', '_logout'])

# Filter out invalid sessions where login happens after logout
df = df[df['timestamp_login'] < df['timestamp_logout']]

# Extract the date from the login timestamp
df['date'] = df['timestamp_login'].dt.floor('d')
```

```python
# Group by user_id and date, then get the max login and min logout for each group
df = df.groupby(['user_id', 'date']).agg(
    timestamp_login=('timestamp_login', 'max'),
    timestamp_logout=('timestamp_logout', 'min')
).reset_index()

# Calculate session duration in seconds
df['duration'] = (df['timestamp_logout'] - df['timestamp_login']).dt.total_seconds()
```

```python
# Calculate the average session duration per user
result = df.groupby('user_id')['duration'].mean().reset_index()

# Rename the duration column
result.rename(columns={'duration': 'average_session_duration'}, inplace=True)

# Display the result
print(result)
```

Q5: MARKETING CAMPAIGN SUCCESS

```python
visits = pd.read_csv("visits.csv")
```

```python
visits.head()
```

```python
visits.info()
```

```python
visits["visit_date"] = pd.to_datetime(visits["visit_date"])
```

```python
# Get the first visit date for each user
first_visit_dates = visits.groupby('user_id')['visit_date'].min().reset_index()
first_visit_dates.rename(columns={'visit_date': 'first_visit_date'}, inplace=True)
```

```python
# Merge the first visit dates with the original visits dataframe
visits = pd.merge(visits, first_visit_dates, on='user_id')

# Filter out visits that occur on the first visit date
df_subsequent_visits = visits[visits['visit_date'] > visits['first_visit_date']]

# Check for unique engagements (new pages visited after the first visit date)
unique_engagements = df_subsequent_visits.groupby('user_id')['page_visited'].nunique().reset_index()
unique_engagements.rename(columns={'page_visited': 'unique_pages_visited'}, inplace=True)

# Count the number of users who engaged with additional content
engaged_users_count = unique_engagements[unique_engagements['unique_pages_visited'] > 0].shape[0]
```

```python
# Print the result
print(f"Number of users engaged with additional content: {engaged_users_count}")
```
